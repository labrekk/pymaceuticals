## Observations and Insights 

```python
# Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress
# Import csv files from resource data folder
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# csv reading
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
mergedf = pd.merge(mouse_metadata, study_results, how='outer', on='Mouse ID')
# Display the data table for preview
mergedf
```

```python
# Checking the number of mice.
num_mice = len(mergedf['Mouse ID'].value_counts())
print(num_mice)
# 249 mice were part of this study.
# Mouse g989 was the only mouse counted >10 times. This signals that data on g989 may contain duplicates. 
mergedf['Mouse ID'].value_counts()
```

```python
# Based on the previous guess of g989 being the dupe, in this cell I am

# Source https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.duplicated.html

# This line of code locates Mouse ID duplicates in the Mouse ID/Timepoint columns of mergedf - all belong to g989.

# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 

mergedf.loc[mergedf.duplicated(subset=['Mouse ID', 'Timepoint',]),'Mouse ID'].unique()
```

```python
# All data for Mouse g989:

# Index on Mouse ID column

mergedf = mergedf.set_index('Mouse ID')

# Locate all g989 rows with newly indexed mergedf
mouse_g989 = mergedf.loc['g989', :]
mouse_g989
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df = mergedf.drop(index = 'g989')

# Came back up to this cell to reset the index after I got an error running the code in the below cell:
clean_df = clean_df.reset_index()

```

```python
# Checking the number of mice in the clean DataFrame.
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.reset_index.html
new_num_mice = len(clean_df['Mouse ID'].value_counts())
print(new_num_mice)
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
regimen_df = clean_df.groupby('Drug Regimen')
# regimen_df.describe()
# mean, median, variance, standard deviation, and SEM of the tumor volume. 


# Create variables to hold each of the summary statistics:
tumorvol_mean = regimen_df['Tumor Volume (mm3)'].mean()
tumorvol_med = regimen_df['Tumor Volume (mm3)'].median()
tumorvol_var = regimen_df['Tumor Volume (mm3)'].var()
tumorvol_std = regimen_df['Tumor Volume (mm3)'].std()
tumorvol_sem = regimen_df['Tumor Volume (mm3)'].sem()
# Assemble the resulting series into a single summary dataframe.
stat_sum = pd.DataFrame({'Mean': tumorvol_mean,
                         'Median': tumorvol_med,
                         'Variance': tumorvol_var,
                         'Standard Deviation': tumorvol_std,
                         'Standard Error': tumorvol_sem                        
                        })
stat_sum
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
stat_sum
# Using the aggregation method, produce the same summary statistics in a single line
#  https://pandas.pydata.org/docs/getting_started/intro_tutorials/06_calculate_statistics.html
agg_stats = regimen_df.agg({'Tumor Volume (mm3)': ['mean', 'median', 'var', 'std', 'sem']    
                           })
agg_stats
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse

# For 'EACH MOUSE' indicates groupby mouse id, but to get timepoint, also groupby timepoint
timepoint_df = pd.DataFrame(clean_df.groupby('Mouse ID')['Timepoint'].max())
timepoint_df.reset_index()

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
new_df = pd.merge(clean_df, timepoint_df, on='Mouse ID')
new_df = new_df.rename(columns={'Mouse ID': "Mouse ID", "Drug Regimen": 'Drug Regimen', "Sex": 'Sex',
                                      "Age_months": "Age_months", "Weight (g)": "Weight (g)", "Timepoint_x": "Timepoint",
                                      'Tumor Volume (mm3)': 'Tumor Volume (mm3)', 'Metastatic Sites': 'Metastatic Sites', 
                                       'Timepoint_y': 'Last Timepoint'
                                      
                                      })
new_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
drugs_tumors = []



for treatment in treatments:
    # Locate the rows which contain mice on each drug and get the tumor volumes
    #Where all the mouses where timepoint = last timepoint for the 4 drugs
    # Notes: I wanted the Regimen == Treatment here so that only the 4 listed drugs are included in the loop.
    # Timepoint == Last Timepoint so we only take the LAST measurement.
    df = new_df.loc[new_df['Drug Regimen'] == treatment]
    timepoints_tumors_df = df.loc[new_df['Timepoint'] == new_df['Last Timepoint']]
    drugs_tumors.append(timepoints_tumors_df['Tumor Volume (mm3)'])
   # Calculate the IQR and quantitatively determine if there are any potential outliers. 

    quartiles = timepoints_tumors_df['Tumor Volume (mm3)'].quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    print(f"The IQR for {treatment} is {iqr}")
    
    # Determine outliers using upper and lower bounds
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    print(f"Tumor volumes below {lower_bound} for {treatment} could be outliers.")
    print(f"Tumor volumes above {upper_bound} for {treatment} could be outliers.")
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
fig1, ax1 = plt.subplots()
ax1.set_title('Final Tumor Volumes')
ax1.set_ylabel('Final Tumor Volume (mm3)')
ax1.boxplot(drugs_tumors)
# https://stats.stackexchange.com/questions/3476/how-to-name-the-ticks-in-a-python-matplotlib-boxplot
plt.xticks([1, 2, 3, 4], ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin'])

plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
# Copied bits from a lot of the code above.

capomulin_only = clean_df.loc[clean_df['Drug Regimen'] == 'Capomulin']
capomulin_only
avg_tumor = pd.DataFrame(clean_df.groupby('Mouse ID')['Tumor Volume (mm3)'].mean())
avg_tumor = avg_tumor.rename(columns={'Tumor Volume (mm3)': "Average Tumor Vol. (mm3)"})
# NOTE on Scatter_df: got an error re. size of dfs I was trying to plot, so I had to merge the avg-tumor one into the original.
scatter_df = pd.merge(capomulin_only, avg_tumor, on='Mouse ID')
scatter_df

plt.scatter(scatter_df['Weight (g)'], scatter_df['Average Tumor Vol. (mm3)'])
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title('Mouse Weight vs Avg. Tumor Volume for Mice on Capomulin')
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
mouse_weight = scatter_df['Weight (g)']
avg_tumor_volume_var = scatter_df['Average Tumor Vol. (mm3)']
correlation = st.pearsonr(mouse_weight, avg_tumor_volume_var)
print(f'the correlation for mouse weight and avg tumor volume for the capomulin regimen is {correlation[0]}')
```

```python
(slope, intercept, rvalue, pvalue, stderr) = linregress(mouse_weight, avg_tumor_volume_var)
regress_values = mouse_weight * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(scatter_df['Weight (g)'], scatter_df['Average Tumor Vol. (mm3)'])
plt.plot(mouse_weight,regress_values,"r-")
plt.annotate(line_eq,(20,37),color="red")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title('Mouse Weight vs Avg. Tumor Volume for Mice on Capomulin')
```
